## Imóveis Residenciais

```python
import pandas as pd
```

```python
dados = pd.read_csv('Dados/aluguel.csv', sep=';')
```

```python
dados.head()
```

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residencial = list(dados['Tipo'].drop_duplicates())
```

```python
residencial
```

```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila']
```

```python
residencial
```

```python
dados.head(10)
```

```python
#determinando o que queremos do DF
selecao = dados['Tipo'].isin(residencial)
selecao
```

```python
#criando um DF com base no filtro gerado anteriormente
dados_residencial = dados[selecao]
```

```python
dados_residencial
```

```python
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
#conferindo o tamanho do novo DF
dados_residencial.shape[0]
```

```python
dados.shape[0]
```

```python
#readequando o index
dados_residencial.index = range(dados_residencial.shape[0])
```

```python
dados_residencial
```

## Exportando a nova base de dados

```python
dados_residencial.to_csv('Dados/aluguel_residencial.csv', sep=';')
```
