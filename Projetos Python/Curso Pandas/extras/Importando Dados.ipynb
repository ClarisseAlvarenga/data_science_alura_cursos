# Importando e visualizando tipos de arquivos diferentes no Pandas

## JSON

```python
#visualizando json
json = open('dados/aluguel.json')
print(json.read())
```

```python
#importando json como DataFrame
df_json = pd.read_json('dados/aluguel.json')
df_json
```

## TXT

```python
#visualizando txt
txt = open('dados/aluguel.txt')
print(txt.read())
```

```python
##importando txt como DataFrame
df_txt = pd.read_table('dados/aluguel.txt')
df_txt
```
