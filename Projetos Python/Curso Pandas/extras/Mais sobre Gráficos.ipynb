```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize=(15,8))
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep=';')
dados.head()
```

```python
area = plt.figure()
```

```python
#colocando 4 gráficos dentro da figura
g1 = area.add_subplot(2,2,1) # informando que são quatro gráficos em duas linhas e duas colunas, e este fica na posição 1
g2 = area.add_subplot(2,2,2) # posição 2
g3 = area.add_subplot(2,2,3) # posição 3
g4 = area.add_subplot(2,2,4) # posição 4
```

```python
#gráfico 1 = ScatterPlot
g1.scatter(dados.Valor, dados.Area)
g1.set_title('Valor X Area')

#gráfico 2 - Histograma
g2.hist(dados.Valor)
g2.set_title('Histograma')

#criando uma amostra aleatória e fazendo um gráfico dela
dados_g3 = dados.Valor.sample(100) #gerando a amostra
dados_g3.index = range(dados_g3.shape[0])#recriando o index
g3.plot(dados_g3)
g3.set_title('Amostra Valor')

#gráfico de barras
#criando um grupo
grupo = dados.groupby('Tipo')['Valor']
#fazendo as labels
label = grupo.mean().index
valores = grupo.mean().values
g4.bar(label,valores)
g4.set_title('Valor Médio por Tipo')

```
