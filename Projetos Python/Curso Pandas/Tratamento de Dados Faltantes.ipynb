## Tratamento de Dados Faltantes

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep=';')
```

```python
dados.head()
```

```python
#método para verificar dados faltantes
dados.isnull()
```

```python
#verificando dados "não faltantes"
dados.notnull()
```

```python
#identifica os dados não nulos por coluna
dados.info()
```

```python
#visualizando o valor nulo dentro do banco de dados
dados[dados['Valor'].isnull()]
```

```python
#eliminar dados no Banco
# Neste caso eliminamos os valores, pois é a variável mais importante dos dados que estamos analisando, 
#afinal não podemos vender imóveis sem valor
A = dados.shape[0]
dados.dropna(subset = ['Valor'], inplace=True) # o inplace altera o banco de dados
B = dados.shape[0]

A - B
```

## Tratamento de dados Faltantes - Continuação

```python
#Tratando condomínio e IPTU
# Condomínio - Excluir condomínio nulo se o Tipo de imóvel for Apartamento
#primeiro passo: Checar quantos valores nulos
dados[dados['Condominio'].isnull()].shape[0]
```

```python
#selecao Imóveis do tipo Apartamento com condomínio nulo
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

```python
#usando a seleção para tirar os valores buscados na seleção (invertendo o booleano da seleção)
A = dados.shape[0]
dados = dados[~selecao] #o til é a negação neste caso
B = dados.shape[0]
A - B
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
#alterando dados faltantes no banco de dados (neste caso para 0)
#dados.fillna(0, inplace = True) um jeito de fazer - altera TODOS os valores faltantes para zero, independente 
#da localização no BD
#Outra forma de fazer, selecionando o tipo da variável:
dados = dados.fillna({'Condominio' :0, 'IPTU' : 0})

```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
dados[dados['IPTU'].isnull()].shape[0]
```

```python
#substituindo o dataFrame
dados.to_csv('dados/aluguel_residencial.csv', sep=';', index = False)
```
