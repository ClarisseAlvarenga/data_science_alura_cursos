## **Aula 03 - Anotações e diversos gráficos**

```
#criando uma área de plotagem
fig = plt.figure(figsize=(15,8))
#declarando os eixos
eixo = fig.add_axes([0,0,1,1]) #lista que cria o eixo
eixo.plot(df['data'], df['temperatura'], color='g')
eixo.set_ylabel('Temperatura K', fontsize=20)
eixo.set_xlabel('Data',fontsize=20);
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)
#criando linhas horizontais na temperatura mínima e máxima da série temporal
eixo.axhline(max(df['temperatura']), color='k', linestyle='--')
eixo.axhline(min(df['temperatura']), color='k', linestyle='--')
#criando uma anotação
x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])
eixo.annotate('Máximo', xy=(x1,y1), fontsize=20);
```

```
#MUDANDO A ANOTAÇÃO DO VALOR MÁXIMO
#criando uma área de plotagem
fig = plt.figure(figsize=(15,8))
#declarando os eixos
eixo = fig.add_axes([0,0,1,1]) #lista que cria o eixo
eixo.plot(df['data'], df['temperatura'], color='g')
eixo.set_ylabel('Temperatura K', fontsize=20)
eixo.set_xlabel('Data',fontsize=20);
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)
#criando linhas horizontais na temperatura mínima e máxima da série temporal
eixo.axhline(max(df['temperatura']), color='k', linestyle='--')
eixo.axhline(min(df['temperatura']), color='k', linestyle='--')
#criando uma anotação
x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])
#deslocando a posição da anotação
x2 = df['data'][df['temperatura'].idxmax()-7000] #por hora
y2 = max(df['temperatura']) - 5
#anotação com texto deslocado e uma seta apontando para o ponto correto
eixo.annotate('Máximo', xy=(x1,y1), fontsize=20, xytext=(x2,y2), arrowprops=dict(facecolor='k'));
```

```
#REPLICANDO PARA O VALOR MÍNIMO
#criando uma área de plotagem
fig = plt.figure(figsize=(15,8))
#declarando os eixos
eixo = fig.add_axes([0,0,1,1]) #lista que cria o eixo
eixo.plot(df['data'], df['temperatura'], color='g')
eixo.set_ylabel('Temperatura K', fontsize=20)
eixo.set_xlabel('Data',fontsize=20);
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)
#criando linhas horizontais na temperatura mínima e máxima da série temporal
eixo.axhline(max(df['temperatura']), color='k', linestyle='--')
eixo.axhline(min(df['temperatura']), color='k', linestyle='--')
#criando uma anotação
x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])
#deslocando a posição da anotação
x2 = df['data'][df['temperatura'].idxmax()-7000] #por hora
y2 = max(df['temperatura']) - 5
#anotação com texto deslocado e uma seta apontando para o ponto correto
eixo.annotate('Máximo', xy=(x1,y1), fontsize=20, xytext=(x2,y2), arrowprops=dict(facecolor='k'));
#replicando para o mínimo
x1 = df['data'][df['temperatura'].idxmin()]
y1 = min(df['temperatura'])
#deslocando a posição da anotação
x2 = df['data'][df['temperatura'].idxmin()-7000] #por hora
y2 = min(df['temperatura']) + 5
#anotação com texto deslocado e uma seta apontando para o ponto correto
eixo.annotate('Mínimo', xy=(x1,y1), fontsize=20, xytext=(x2,y2), arrowprops=dict(facecolor='k'));
```

```
#agrupando por dia da semana e aplicando a média na temperatura
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
#criando um filtro para classificar os dados
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
#filtrando por dia da semana
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[nome_dias]
temperatura_por_dia_da_semana
```

```
#criando o gráfico
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
indice = range(len(temperatura_por_dia_da_semana))
eixo.bar(indice,temperatura_por_dia_da_semana)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura', fontsize=15)
```

```
#alterando os ticks
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
indice = range(len(temperatura_por_dia_da_semana))
eixo.bar(indice,temperatura_por_dia_da_semana)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura', fontsize=15)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias);
```

```
#incluindo cores
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
cores = ['k', 'r', 'g', 'b', 'yellow', 'orange', 'magenta']
indice = range(len(temperatura_por_dia_da_semana))
eixo.bar(indice,temperatura_por_dia_da_semana, color=cores)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura', fontsize=15)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias);
```

```
#incluindo bordas nas colunas
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
cores = ['k', 'r', 'g', 'b', 'yellow', 'orange', 'magenta']
indice = range(len(temperatura_por_dia_da_semana))
eixo.bar(indice,temperatura_por_dia_da_semana, color=cores, edgecolor='k')
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura', fontsize=15)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias);
```

## **Aula 05 - Customizando Histogramas**

```
#anotando a média e desvio padrão
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

mu,sigma = iris['comprimento_pétala'].mean(),iris['comprimento_pétala'].std()

eixo.hist(iris['comprimento_pétala'], bins=20, )
eixo.set_title('Histograma', fontsize=15, pad=10)
eixo.set_ylabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True);
#os cifrões indicam que vem uma fórmula
eixo.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(mu,sigma),
              xy=(4.5,20), fontsize=20);
```

```
#incluindo linhas verticais para a média e a mediana
fig = plt.figure(figsize=(8,5))
eixo = fig.add_axes([0,0,1,1])

mu,sigma = iris['comprimento_pétala'].mean(),iris['comprimento_pétala'].std()

eixo.hist(iris['comprimento_pétala'], bins=20, )
eixo.set_title('Histograma', fontsize=30, pad=10)
eixo.set_ylabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True);
#os cifrões indicam que vem uma fórmula
eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu,sigma),
              xy=(4.5,20), fontsize=20);
#incluindo linha e anotação média
eixo.axvline(mu, color= 'k', linestyle='--')
eixo.annotate('média', xy=(mu-0.79,28), fontsize=20);
#incluindo linha e anotação mediana
eixo.axvline(iris['comprimento_pétala'].median(), color= 'g', linestyle='--')
eixo.annotate('mediana', xy=(iris['comprimento_pétala'].median(),31), fontsize=20);
```
